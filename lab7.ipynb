```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Функція для побудови графа Y-комбінатора
def draw_y_combinator():
    # Створюємо порожній граф
    G = nx.DiGraph()
    # Додаємо вузли для лямбда-абстракцій і функцій
    G.add_node("λf")
    G.add_node("λx (x x)")
    G.add_node("f (x x)")
    G.add_node("(λx (x x)) (λx (x x))")
    
    # Додаємо ребра, які представляють застосування функцій
    G.add_edges_from([("λf", "(λx (x x)) (λx (x x))"),
                      ("λx (x x)", "f (x x)"),
                      ("f (x x)", "(λx (x x)) (λx (x x))")])
    # Візуалізація графа
    pos = nx.spring_layout(G)  # Позиція для графа
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold')
    plt.title("Y-комбінатор Каррі")
    plt.show()
```

```python
# Викликаємо функцію для побудови графа
draw_y_combinator()
```
